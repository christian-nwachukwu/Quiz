```python
question_number = 0

score = 0

QUESTIONS = (
    "When was the first known use of the word 'quiz'? :",
    "Which built-in function can get information from the user? :",
    "Which keyword do you use to loop over a given list of elements? :",
    "What's the purpose of the built-in zip() function? :",
    "What's the name of Python's sorting algorithm? :",
    )

CHOICES = (
    (
        "A. 1781",
        "B. 1771",
        "C. 1871",
        "D. 1881",
    ),

    (
        "A. input",
        "B. get",
        "C. print",
        "D. write",
    ),
    (
        "A. for",
        "B. while",
        "C. each",
        "D. loop",
    ),
    (
        "A. To iterate over two or more sequences at the same time",
        "B. To combine several strings into one",
        "C. To compress several files into one archive",
        "D. To get information from the user",
    ),
    (
        "A. Timsort",
        "B. Quicksort",
        "C. Merge sort",
        "D. Bubble sort",
    )
)

ANSWERS = ("A", "A", "A", "C", "A")

guesses = []

# Input possible answer
def response():
    guess = input("Please select an answer(A, B, C, or D): ").upper()
    guesses.append(guess)
    return guess

for question in QUESTIONS:
    print(question)
    print()
    print('.......................................')
    for choice in CHOICES[question_number]:
        print(choice)
        print()

    # Calling the response to question function
    guess = response()

    if guess == ANSWERS[question_number]:
        score += 1
        print("Correct!")
        print()
    else:
        print("INCORRECT!")
        print()


    question_number += 1


print("answer :", end="")
for answer in ANSWERS:
    print(answer , end="")
print()


print("guess :", end="")
for guess in guesses:
    print(guess, end="")
print()
print()
print()


# Result
print("....................RESULT........................")
score = int(score / len(QUESTIONS) * 100)
print(f"Your score is {score}%")



def declare_result(score):
    if score >= int(50):
        print("You PASSED, CONGRATULATIONS!!!")
    else:
        print("You FAILED, TRY AGAIN!!!")

declare_result(score)
```
